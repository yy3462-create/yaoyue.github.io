# Repeat the previous step using only the Python standard library

```python
# here is the solution using only the Python standard library (rather than simply import pandas, I need to import csv and use csv.reader)
import csv

values = []

with open("Small_Business_Administration_(SBA)_Size_Standards_Table_20251107.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        raw_val = row[col]
        if raw_val:
            cleaned = ''.join(ch for ch in raw_val if ch.isdigit() or ch == '.')
            if cleaned:
                values.append(float(cleaned))
print("Values:", values[:50]) # print first 50 cleaned values
```

```python
mean_val = sum(values) / len(values)
sorted_vals = sorted(values)
n = len(sorted_vals)
if n % 2 == 1:  # if the number of values is odd
    median_val = sorted_vals[n // 2]
else:  # if the number of values is even
    median_val = (sorted_vals[n // 2 - 1] + sorted_vals[n // 2]) / 2\

freq = {}  
for v in values:
    freq[v] = freq.get(v, 0) + 1
mode_val = max(freq, key=freq.get)     # get the value with the highest frequency

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
    
```

# Data visualization

```python
import csv
import random
csv_path = "Small_Business_Administration_(SBA)_Size_Standards_Table_20251107.csv"

col = "Size Standards in millions of dollars"
label_col = "NAICS Industry Description"


# read and clean data
data = []
# first, read the CSV file
with open(csv_path, newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    # then, extract and clean the relevant columns
    for row in reader:
        label = row[label_col]
        raw_val = row[col]
        # clean the numeric value, only keep digits and decimal points
        if label and raw_val:
            cleaned = ''.join(ch for ch in raw_val if ch.isdigit() or ch == '.')
            # convert to float if cleaned value is not empty
            if cleaned:
                data.append((label, float(cleaned)))
print (data[:5])  # print first 5 entries to verify

```

```python
# we can randomly sample 5 entries for visualization
random.seed(42)
sample = random.sample(data, 5)

# determine scaling factor
max_val = max(v for _, v in sample)
scale = 25 / max_val  # control max length to 25 symbols

print("SBA Size Standards by Random Industries (in millions of $)")
print("Each ▇ ≈ proportional to value\n")
print(f"{'Industry':35} | {'Visualization':26} | Value")
print("-" * 70)

# visualize each entry
for label, val in sample:
    bar = "▇" * int(val * scale)
    print(f"{label[:33]:35} | {bar:<26} | {val:>6.2f}") # align label to left, value to right
```
